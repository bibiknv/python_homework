```python
# https://stepik.org/lesson/520159/step/17?unit=512678

import sys
from datetime import datetime

MODE='TEST'

if MODE=='TEST':
    fin = open('test.txt', encoding='utf')
    fout = open('test_out.txt', 'w')
else:
    fin = sys.stdin
    fout = sys.stdout
    
with fin, fout:
    list_dates = []
    for line in fin:
        line = line.strip()
        datetime_object = datetime.strptime(line, "%d.%m.%Y")
        list_dates.append(datetime_object)
    
    sorted_list_dates_true = sorted(list_dates, reverse = True)
    sorted_list_dates_false = sorted(list_dates)
    set_list_dates = list(set(list_dates))
    
    if len(set_list_dates) != len(list_dates):
        fout.write('MIX')
    elif len(set_list_dates) == len(list_dates):
        if list_dates == sorted_list_dates_false:
            fout.write('ASC')
        elif list_dates == sorted_list_dates_true:
            fout.write('DESC')
        else:
            fout.write('MIX')
    
```

```python
import sys


def  mytest(lines):
    delta=lines[1]-lines[0]
    if all([lines[i]==lines[0]+delta*i for i in range(len(lines))]):
        return ('Арифметическая прогрессия') 
    try:
        delta=lines[1]/lines[0]
        if all(lines[i]==lines[0]*delta**i for i in range(len(lines))):
            return ('Геометрическая прогрессия') 
    except: 
        pass
    
    return ('Не прогрессия')
    
    
    

#sys.stdin=open('input.txt')
lines = [int(line) for line in sys.stdin]
print(mytest(lines))

```
