```python
# https://stepik.org/lesson/518491/step/13?unit=510939

import csv

with open('salary_data.csv', 'r', encoding='utf-8') as csv_file:
    next(csv_file)
    d = {}
    count = 0
    data = csv_file.read()
    table = [r.split(';') for r in data.splitlines()]
    for company, salary in table:
        d.setdefault(company, {'sum': 0, 'count': 0})
        d[company]['sum'] += int(salary)
        d[company]['count'] += 1

    for company in d:
        d[company]['avg'] = d[company]['sum'] / d[company]['count']
    
    items = d.items()
    sorted_dict  = sorted(items, key =lambda x: x[1]['avg'])
    
    for i in sorted_dict:
        print(i[0])
        
```

```python
import csv

d = {}

with open('salary_data.csv', encoding='utf-8') as file:
    rows = list(csv.reader(file, delimiter=';'))
    for key, value in rows[1:]:
        d[key] = d.get(key, []) + [int(value)]

    d_sort = sorted(d, key=lambda x: (sum(d[x]) / len(d[x]), x))
    print(*d_sort, sep='\n')
```
