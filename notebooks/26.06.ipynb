```python
import csv

data = [{'first_col':'value1', 'second_col': 'value2'}]
columns = ['first_col', 'second_col']

with open('writing_test.csv', 'w', encoding='utf-8', newline = '') as file:
    # создаем writer объект и указываем названия столбцов
    writer = csv.DictWriter(file, fieldnames = columns, delimiter=',')
    writer.writeheader()
    for row in data:
        writer.writerow(row)
```

```python

import csv

with open('writing_test.csv', 'w', encoding='utf-8') as csv_file:
    keys, data = ['first_col', 'second_col'], ['value1', 'value2']
    writer = csv.DictWriter(csv_file, fieldnames=keys)
    writer.writeheader()
    writer.writerow(dict(zip(keys, data)))
```

```python
import csv

with open('sales.csv', 'r', encoding='utf-8') as csv_file:
    rows = csv.reader(csv_file, delimiter=';', quotechar='"')
    for i, row in enumerate(rows):
        if i > 0 and int(row[2]) > int(row[1]):
            print(row[0])
```
